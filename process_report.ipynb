```python
import pandas  as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_table('report.tsv',header=None)
```

```python
df.columns=['Reference genes','Reference sequence lenght','Mapped read count','Unmaped read count']
#df=df[['Reference genes','Mapped read count']]
```

```python
df.apply(pd.to_numeric, errors='ignore')
```

```python
df['Percentage of Mapped reads']=(df['Mapped read count']*100)/df['Mapped read count'].sum()
```

```python
df[df['Percentage of Mapped reads'] > 0].plot(kind='barh',x='Reference genes',y='Percentage of Mapped reads',figsize=(5,10))
plt.xlabel('Percentage of Mapped reads')
```
